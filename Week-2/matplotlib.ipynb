```python
import matplotlib.pyplot as plt
```

Basic Plot (Line Plot)


```python
x = [1, 2, 3, 4, 5]
y = [2, 5, 7, 1, 6]

plt.plot(x, y)          # make the line plot
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Bar Chart

```python
students = ["Ali", "Sara", "John", "Aisha"]
marks = [85, 90, 78, 88]

plt.bar(students, marks, color="skyblue")
plt.title("Students' Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

Scatter Plot

```python
mileage = [127300, 57850, 39190, 210814, 127869]
price   = [3074, 49704, 24072, 1705, 4101]

plt.scatter(mileage, price, color="red")
plt.title("Car Price vs Mileage")
plt.xlabel("Mileage")
plt.ylabel("Price")
plt.show()
```

Histogram
shows frq distribution


```python
ages = [18, 20, 17, 19, 18, 21, 20, 19, 18, 17]

plt.hist(ages, bins=5, color="green", edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

Customization

```python
plt.plot(x, y, label="Line 1")
plt.plot(x, [i*2 for i in y], label="Line 2")
plt.legend()            #a small box on the graph that tells which line is Line 1 and which is Line 2
plt.show()
```

Car DataSet


```python
import pandas as pd
df = pd.read_csv("car_sales_data.csv")

# Average price by Manufacturer (bar chart)
avg_price = df.groupby("Manufacturer")["Price"].mean()

avg_price.plot(kind="bar", color="orange")
plt.title("Average Car Price by Manufacturer")
plt.ylabel("Price")
plt.show()
```

```python
# Count cars by fuel type
fuel_counts = df["Fuel type"].value_counts()
print(fuel_counts)

# Make the pie chart
plt.pie(
    fuel_counts, 
    labels=fuel_counts.index,        # names (Petrol, Diesel, Hybrid)
    autopct='%1.1f%%',               # show percentages
    startangle=90,                   # rotate chart to start at the top
    colors=["gold", "skyblue", "lightgreen"]  # optional colors
)

plt.title("Distribution of Cars by Fuel Type")
plt.show()
```
